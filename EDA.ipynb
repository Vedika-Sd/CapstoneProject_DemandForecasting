# AI driven Festival Aware Demand Forecasting and decision support systems for Dairy Manufacturing

The raw Krishna Dairy dataset (ALL_YEARS_COMBINED.csv) was cleaned in a structured pipeline. First, all string columns were stripped of extra whitespace, and four irrelevant columns — DC_NO, SUB_POS_NAME, GSTIN, and HSN — were dropped as they were not needed for analysis. The core challenge was the PRODDESC column, which contained 46 unique product names garbled in Akruti font encoding, which were translated to clean English names using a master mapping dictionary with 4-level fuzzy matching. The DC_DATE column was parsed from mixed date formats into a standardized MM-DD-YYYY datetime format and sorted chronologically, making the data fully ready for time series forecasting models like ARIMA, Prophet, or LSTM. Finally, fully duplicate rows were removed and the cleaned dataset was saved as ALL_YEARS_CLEANED.csv.
